A Calabouço Discos é uma loja de vinis que resiste ao digital com albuns clássicos de lendas do rock. Os albuns são catalogados por: id, ano, artista, genero, titulo, preço. Os clientes são armazenados com: id, nome e idade. E as vendas da loja são armazenadas da seguinte maneira: id, id_cliente, ids_albuns (que é uma lista de IDs com os albuns que a pessoa comprou), total_da_venda.

 Utilizando dicionários, crie um código para registrar as vendas e ao final apresentar o fechamento do dia, apresentando as vendas uma a uma.

* Plus1: Quantas vendas foram realizadas? - Ok
* Plus2: Qual foi o faturamento total da loja? - Ok
* Plus3: Qual cliente mais gastou? - Ok
* Plus4: Qual foi a maior compra? - Ok
* Plus5: Qual é o album mais vendido? - Ok
* Plus6: Quais sao os generos mais comprados pelas faixas de idade abaixo:
  - 18 a 25 anos
  - 25 a 35 anos
  - 35 a 55 anos
  - acima de 55 anos


```python
class Clients : 

    clients = {
        "001": {"nome": "João", "idade": 15},
        "002": {"nome": "Maria", "idade": 17},
        "003": {"nome": "José", "idade": 24},
        "004": {"nome": "Clarêncio", "idade": 34},
        "005": {"nome": "Juvenal", "idade": 75},
        "006": {"nome": "Teotônio", "idade": 54},
        "007": {"nome": "Olandina", "idade": 39},
    }

class Albuns : 

    albuns = [
    {
        "id": "001",
        "ano": 2007,
        "artista": "New Found Glory",
        "genero": "Rock",
        "titulo": "Makes Me Sick",
        "preco": 38.22,
    },
    {
        "id": "002",
        "ano": 1956,
        "artista": "Miles Davis",
        "genero": "Jazz",
        "titulo": "Kind of Blue",
        "preco": 91.64,
    },
    {
        "id": "003",
        "ano": 2012,
        "artista": "Madonna",
        "genero": "Pop",
        "titulo": "Like a Virgin",
        "preco": 48.46,
    },
    {
        "id": "004",
        "ano": 1997,
        "artista": "Foo Fighters",
        "genero": "Rock",
        "titulo": "Learn to Fly",
        "preco": 35.58,
    },
    {
        "id": "005",
        "ano": 1996,
        "artista": "Metallica",
        "genero": "Rock",
        "titulo": "Master of Puppets",
        "preco": 77.21,
    },
]

```

```python
class Vendas(Clients, Albuns):
    """
    Classe que representa as vendas de álbuns para os clientes.

    A classe herda as classes Clients e Albuns para utilizar os dicionários dentro dela.
    """

    counter = 0
    sales = list()
    verifier_age = {
    "0-18"  : range(0,18),
    "18-25" : range(18, 25),
    "25-35" : range(25, 35),
    "35-55" : range(35, 55),
    "55+"   : range(55, 120)
    }

    @classmethod
    def incrementa_contador(cls):
        """
        Método de classe para incrementar o contador de vendas realizadas.
        """
        cls.counter += 1

    # Método que realiza a venda
    def realiza_venda(self):
        """
        Método para realizar uma venda.

        Solicita o ID do cliente e o ID do álbum para a venda.
        Permite adicionar múltiplos álbuns à venda.
        Calcula o preço total da venda.
        Adiciona a venda à lista de vendas realizadas.
        """
        id_client = ""
        self.incrementa_contador()
        sale = list()
        while not id_client.isdigit():
            id_client = input("Digite o ID do Cliente: ")

        while True:
            id_album = input("Digite o ID do Álbum: ")
            option = input("Digite 1 - Pra continuar ou 2 - Parar ")
            if option.isdigit() and int(option) == 1:
                sale.append(id_album)
                continue
            elif option.isdigit() and int(option) == 2:
                sale.append(id_album)
                break

        # Calcula o preço total de cada venda
        total_price = 0
        for id_sale in sale:
            for album in Albuns.albuns:
                if id_sale == album["id"]:
                    total_price += album["preco"]

        # Retorna uma lista com cada venda
        return self.sales.append(
            {
                "id_sale": f"0{self.counter}",
                "id_client": id_client,
                "id_albuns": sale,
                "total_price": total_price,
            }
        )

    @classmethod
    def quantidade_vendas(cls):
        """
        Método de classe que retorna a quantidade de vendas realizadas.
        """
        return f"Foram efetuadas {len(cls.sales)} Vendas!"

    @classmethod
    def faturamento_total(cls):
        """
        Método de classe que retorna o faturamento total das vendas.
        """
        return f"O faturamento total foi : R$ {sum([elem['total_price'] for elem in cls.sales])}"

    def report_venda(self, maior_compra: bool = False):
        """
        Método de classe que retorna o cliente que mais gastou.
        """
        major_cost = 0
        for elem in self.sales:
            if elem["total_price"] > major_cost:
                id_client = elem["id_client"]
                major_cost = elem["total_price"]

        if maior_compra:
            for id_ in Clients.clients:
                if id_client == id_:
                    for elem in self.sales:
                        if id_client == elem["id_client"]:
                            albuns = [
                                album["titulo"]
                                for album in Albuns.albuns
                                for id_album in elem["id_albuns"]
                                if id_album == album["id"]
                            ]
                            print(
                                f"A maior compra foi efetuada por : {Clients.clients[id_]['nome']}\nOnde foram comprados os álbuns : {albuns}\nNo valor de: {major_cost}"
                            )

        for id_ in Clients.clients:
            if id_client == id_:
                return f"Cliente que mais gastou : {Clients.clients[id_]['nome']}"
    
    def album_mais_vendido (self, top:int=5) :

        best_sellers = dict()
        for elem in self.sales :

            for id_album in elem["id_albuns"] :

                best_sellers[id_album] = best_sellers.get(id_album, 0) + 1
        
        
        sorted_best_sellers = dict(sorted(best_sellers.items(), key= lambda x: x[1], reverse=True))

        for index, id_album in enumerate(list(sorted_best_sellers.keys())[:top]) :

            for elem in Albuns.albuns :

                if id_album == elem["id"] :

                    print(
                        f"O {index + 1 }º álbum mais vendido foi: {elem['titulo']} com {sorted_best_sellers.get(id_album)} cópias."
                    )
    
    @classmethod
    def most_sale_per_age (cls) : 
        dict_ = dict()
        for elem in cls.sales : 

            id_client = elem["id_client"]
            id_albuns = elem["id_albuns"]

            idade_client = Clients.clients[id_client]["idade"]

            for k in cls.verifier_age :
                range_ = cls.verifier_age].get(k, 0)
                if idade_client in range_ :
                    for id in id_albuns :
                        for album in Albuns.albuns :
                            if id == album["id"] :
                                genero = album["genero"]
                                dict_.setdefault(k, {genero: 0})
                                dict_[k][genero] = dict_[k].get(genero, 0) + 1
        
        return dict_
    


```

```python
vendas_handler = Vendas()
```

```python
for i in range(5) :
    vendas_handler.realiza_venda()
```

```python
vendas_handler.report_venda(maior_compra=True)
```

```python
vendas_handler.album_mais_vendido(top=2)
```

```python
vendas_handler.most_sale_per_age()
```
